```python
import wget
base_url = 'http://hgdownload.cse.ucsc.edu/goldenPath/hg19/encodeDCC/wgEncodeAwgTfbsUniform'
filename = wget.download(f'{base_url}/files.txt', out='dataset')
```

```python
with open('dataset/files.txt') as txt:
  lines = txt.readlines()
  for line in lines:
    filename = line.split()[0]
    wget.download(f'{base_url}/{filename}', out='dataset')
```
